# 项目：评估和清理英国电商公司销售数据

## 分析目标

此数据分析的目的是，根据市场销售数据，挖掘畅销产品，以便制定更有效的市场策略来提升营收。

本实战项目的目的在于练习评估数据干净和整洁度，并且基于评估结果，对数据进行清洗，从而得到可供下一步分析的数据。

## 简介

原始数据集记录了一家英国在线零售公司在2010年12月1日至2011年12月9日期间的所有交易情况，涵盖了该公司在全球不同国家和地区的业务数据。该公司主要销售覆盖各个场景的礼品，包括但不限于生日礼品、结婚纪念品、圣诞礼品等等。该公司的客户群体主要包括批发商和个人消费者，其中批发商占据了相当大的比例。

数据每列的含义如下：
- `InvoiceNo`: 发票号码。6位数，作为交易的唯一标识符。如果这个代码以字母“c”开头，表示这笔交易被取消。
- `StockCode`: 产品代码。5位数，作为产品的唯一标识符。
- `Description`: 产品名称。
- `Quantity`: 产品在交易中的数量。
- `InvoiceDate`: 发票日期和时间。交易发生的日期和时间。
- `UnitPrice`: 单价。价格单位为英镑（£）。
- `CustomerID`: 客户编号。5位数，作为客户的唯一标识符。
- `Country`: 国家名称。客户所居住的国家的名称。

## 评估数据干净度

```python
product_df.info()
```

通过info得到表格概况，发现`在Description` 和`CustomerID`两列中存在NAN值。  
在数据类型方面，InvoiceDate应该为date，CustomerID应该为str,应当进行格式转换

可能存在问题：  
1. 缺失数据
2. 重复数据
3. 不一致数据
4. 无效数据
5. 错误数据  
逐一判断

### 1.筛选缺失数据

```python
#筛选description中为缺失值的行
product_df[product_df["Description"].isnull()]
```

这些description为NaN的数据，unitprice都为0.0,添加筛选条件查找description为NaN但unitprice不为0的值

```python
check_unit = product_df[product_df["Description"].isnull()]
check_unit[check_unit["UnitPrice"] != 0]
# #或者
product_df[product_df["Description"].isnull() & product_df["UnitPrice"] != 0.0]
```

以上output为空，说明所有description为空的值，price也都为0  
这两个值同时缺失导致这些信息无效，需要删除

```python
#提取customer id的缺失值
product_df[product_df["CustomerID"].isnull()]
```

因为customerid对分析过程无影响，所以保留

### 2.评估重复值

通过观察，表格中的数据并不存在只能出现一次的数据类型，即使一些数据为唯一指定（交易商品客户id）也可以反复出现。
所以无需评估重复值

### 3. 评估不一致数据

查看不同country，是否存在不同str表示同一国家情况

```python

product_df["Country"].value_counts()
```

国家列表中，USA和United States都表示美国，而U.K.，UK，united kingdom都表示英国，需要统一

### 评估无效和错误数据

通过describe对数值统计信息进行了解

```python
product_df.describe()
```

quantity 和price出现负数，需要检查


```python
product_df[product_df["Quantity"] < 0]
```

提出猜测，所有Quantity为负数的交易都是取消交易的，进行进一步筛选

```python
df2 = product_df[(product_df["Quantity"] < 0) & (product_df["InvoiceNo"].str[0] != "C")]
df2
```

有不以C开头的数据，且Quantit为负数，说明并不是所有的负数Quantity订单都被取消了。继续观察所有的price都是0，继续筛选

```python
df2[(df2["UnitPrice"] != 0)]
```

筛选出条件为空，所以说明如果Quantity为负数，该交易要不然被取消了，要不然价格收益为0  
这些数据不为有效数据，没有实际交易，没有产生实际收益，不在分析范畴之内，需要删除

```python
#筛选unit price
product_df[product_df["UnitPrice"] < 0]
```

只有两行数据，而且为Adjust bad debt并非交易数据，对分析无用，需要删除

## 清理数据

整理评估结果，需要处理：  
* 数据类型转换，date和ID
* 删除所有`description`为NaN的行
* 统一国家名称（US，UK）
* 删除Quantity和Price为负数的数据

```python
#将时间从str改成datetim
new = product_df
new
```

```python
new["InvoiceDate"] = pd.to_datetime(new["InvoiceDate"])
```

```python
new.info()
```

```python
new["CustomerID"] = new["CustomerID"].astype(str)
new.info()
```

```python
new.head()
```

出现新问题：保存后的ID作为str依然有0.0，需要删除小数位

```python
new["CustomerID"] = new["CustomerID"].str.slice(0,5)
new.head()
```

```python
#删除缺失值
new = new.dropna(subset = ["Description"]) 
```

```python
new["Description"].isnull().sum()
```

```python
new["Country"] = new["Country"].replace({"USA":"United State"})
```

```python
new["Country"] = new["Country"].replace({"U.K.":"United Kingdom","UK":"United Kingdom"})
```

```python
new["Country"].value_counts()
```

```python
#删除负数
new_1 = new[new["Quantity"] >0]
new_2 = new_1[new_1["UnitPrice"] > 0]
new_2
```

```python
len(new_2[new_2["Quantity"] < 0])
```

并不是进行删除负数行，而是将非负数行筛选出即可
